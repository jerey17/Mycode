## Jupyter Lab

- **Recall:** We will be using a notebook system called `Jupyter`. 

- **Reference:** https://jupyterlab.readthedocs.io/en/latest/
    - Also, within the JupyterLab space, Help --> JupyterLab Reference

- Jupyter notebooks allow you to write **cells of code** and **display their output** as text, images, or interactive widgets, as well as **formatted text descriptions**.
    - Excels in reproducibility
    - Easy to share and publish on-line
    
    
- `Jupyter` runs as a **locally hosted web-app** that you interact with through a web browser. 


- `Jupyter` manages running a language-specific [**kernel**](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/What%20is%20the%20Jupyter%20Notebook.html#Kernels) that executes your code and maintains a global variable state. This means that:
    - Variable and function definitions persist between cells.
    - Cells can be run in **any order**.
    - Running a cell will make a **persistent change** to the global state of your interactive session.
    
    
- Notebooks are an excellent space to test out ideas with **rapid iteration**.
    - This is highly compatible with **learning** a programming language through experimentation and **data science** workflows.
    - No need to execute entire programs repeatedly.
    - But you can, if you want to.
    - You can also **convert** notebooks into Python scripts when you are happy with the code.
    

**As best practice, write your notebooks to be run sequentially from top to bottom!**

#### Difficulty Loading Jupyter Notebooks

Some students have experienced difficulty launching JupyterLab on their personal machines.

Launching from within the Anaconda Navigator, from an Anaconda prompt (command: `jupyter lab`), or by selecting Jupyter Notebook from the Windows Start menu all failed to **automatically launch** the Notebook service in a new browser tab. Instead, the latter two options will **present text in a terminal window**, enabling the user to copy and paste a given URL (with token) into the browser; doing this does successfully initiate an active session but is a bit cumbersome.

For example,

```bash
...
[I 2024-10-30 17:36:28.176 ServerApp] http://localhost:8889/lab?token=5ed4df1e0fbdec96365bc71c6cd3f769dc07d97d7828d8bc
[I 2024-10-30 17:36:28.177 ServerApp] http://127.0.0.1:8889/lab?token=5ed4df1e0fbdec96365bc71c6cd3f769dc07d97d7828d8bc
...
```

Researching this issue in the past, I have found that many other users report having encountered this issue recently, and most proposed solutions match our working solution (copy/paste URL from terminal). The reason for this problem might be related to the use of a local system username that includes unicode characters (e.g., Cyrillic or Chinese characters) which are not properly handled by the system interaction with the launching service. Other possible issues relate to errors in installation of or updates to Anaconda and a variety of potential conflicting system path configurations.

We are unlikely to provide a universal solution to this problem. Clever Googling might be the remedy of choice.

#### About Ports

Notebooks are run through **servers** on a local port.  Typically, they try to initially use port `8888` but will try others (typically, sequentially increasing values) if `8888` is in use.  By default, a notebook server runs locally at `127.0.0.1:8888` and is accessible only from localhost.  You may access the notebook server from the browser using http://127.0.0.1:8888. 

When launching from the Navigator, port information is found in the address.

In a prompt or terminal:

- List active jupyter sessions with their tokens: 
```
jupyter server list
```

- Launch with a specific port number:
```
jupyter lab --port <port_number>
```

- Kill active jupyter sessions by specifying the port number: 
```
jupyter server stop <port_number>
```

### `conda` vs `pip`

Similar, but certainly not identical, utilities to manage packages of code. Often have overlapping functionality and can be used in tandem, though this can lead to problems.

- `conda`: a cross platform package and environment manager that installs and manages conda packages from the Anaconda repository as well as from the Anaconda cloud.
    - Binary installation, no need to compile
    - Manages packages from other languages (C, C++, R, etc.)
    - Built-in support for separate environments
    - Robust dependency resolution
    - Over **1,500 packages** on the repository, plus thousands more on the cloud
    - Often requires **wildly unreasonable wait times** to create environments, reconcile version conflicts, and install packages.
    
    
- `pip`: the Python Packaging Authoritys recommended tool for installing packages from the Python Package Index, PyPI.
    - May require further compilation (need appropriate compilers/libraries)
    - Requires use of other (easily installable) tools to manage environments
    - Simple dependency resolution (might lead to subtly broken package combinations)
    - Over **150,000 packages** through PyPI
        - These are not thoroughly inspected; use at your own risk.
    - Typically, installations are **very quick**.

### Creating a new `conda` environment

One example of a potential conflict with our `base` environment arises with the Tensorflow library, which is used in machine learning applications.  In particular, to harness the power of the NVIDIA GPUs on these lab PCs, the library needs to be installed alongside a particular set of dependencies.  The `base` environment only supports CPU computations, and Windows machines have a problem with running GPU enabled Tensorflow in the newest versions of the library (beyond 2.10).  Because this is known to work with an older version of Python, we can specify this as part of the new environment, too.

To account for these issues, we can create a new environment for GPU-enabled Tensorflow.

**Note**: This will take up 4-5 GB. 

Open the Anaconda Prompt. 

Enter the following commands, confirming `y` when prompted.Some of these will take a minute to complete. 
```bash
conda create --name c39tf python=3.9 
conda activate c39tf 
conda install -c conda-forge cudatoolkit=11.2 cudnn=8.1.0 
python -m pip install --upgrade pip 
pip install "numpy<2.0" "tensorflow<2.11" 
```

In order, we:
1. Create the new environment, linked to Python 3.9.
2. Activate the new environment (enter its space).
3. Install NVIDIA support tools in the new environment.
4. Install `pip` in the new environment, so that it is consistent with Python 3.9.
5. **Takes a minute**.  Install tensorflow utilities in the new environment.


When you `activate` the new environment, the prompt prefix is updated to reflect the change.

Try running `conda list` on this environment.  Note that it has many fewer packages than the `base` environment.

Try running `conda env list`.  What does this tell you?

---

Test the installation from the command line: 

```bash
(c39tf) > python 
```
```python
import tensorflow as tf 
tf.config.list_physical_devices() 
```


This should return information about both a CPU and a GPU `PhysicalDevice`, confirming that it indeed accessible by this configuration.In the `base` environment or any other in which GPU support has not been installed, only the CPU information will be returned. 

**If nothing (i.e., `None`) is returned, please report this to your instructor, as there may be problems with the GPU configuration on your machine.**
