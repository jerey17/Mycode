````python
import matplotlib.pyplot as plt
import numpy as np


def plot_line(x, y):
    plt.plot(x, y, label='data')
    plt.title("Sample Line Graph")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")

    yarr = np.asarray(y)
    
    plt.hlines(yarr.mean(), min(x), max(x), color='r', label='mean')
    plt.legend()


x_values = [0, 1, 2, 3,  4,  5]
y_values = [0, 1, 4, 9, 16, 25]

plot_line(x_values, y_values)
````

#### Running Python

- When Python runs a `.py` file, it is first compiled into **byte code**, and a cached version is saved.
    - Byte code is platform agnostic.
    - This process is largely **invisible** to the user/programmer.
    - Translating the code to byte code before running speeds up execution


- There is no separation between compiling and running code.
    - You can simply write a program and execute it. 
    - You can even execute strings containing Python code within a Python program!
    

---
#### Executing Python Scripts

```bash
$ cat hellomars.py 
import sys

with open(sys.argv[1], 'w') as file:
    file.write('Hello, Mars!\n')


$ python hellomars.py output.txt 

$ cat output.txt 
Hello, Mars!
```

<h2 style="color:Crimson;">Indentation and Comments</h2>

**In Python, indentation acts as *blocks* do in `C` or `Java`:**

C/Java:

```java
    if(x<y)
    {
      z = x;
    }
    else
      {
      if(x==y)
      {
        y = y + 1;
      }
      else
      {
        z = y;
      }
    }
```

Python:

```python
    if x < y:
        z = x
    else:
        if x == y:
            y = y + 1
        else:
            z = y
```

- Block **indentation** is **4 spaces**.
    - With Jupyter notebook, in code and markdown cells, the `Tab` key should automatically provide 4 spaces.
    - When selecting multiple lines, `Tab` will shift the entire selection to the right.  
    - `Shift + Tab` will shift the selection to the left.

    
    
- Sometimes, the lack of explicit **block termination** markers can make it difficult to determine which parts of the code belong to which block.
    - As a work-around, you might see a **comment** placed at the end of the block, with matching indentation.
 

---
    
- **Comments** can be written in two ways:
    - `#`: for a single line or in-line
        - Any selected or current line of code can be commented with the `Ctrl + /` or ` + /` shortcut.
    - `''' '''` (triple-quotes, single or double): multi-line "**docstrings**"

<h3 style="color:Crimson;">Variable Names</h3>

A variable can have a **short name** (like x and y) or a more **descriptive name** (age, carname, total_volume). 


- **Rules** for Python variables:
    - A variable name **must** start with a letter or the underscore character.
    - A variable name **cannot** start with a number.
    - A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).
    - Variable names are **case-sensitive** (age, Age and AGE are three different variables).

````python
# Some legal variable names
myname = 'Python'
my_name = 'Python'
_my_name = 'Python'
myName = 'Python'
MYNAME = 'Python'
myname2 = 'Python'
````

````python
# Some illegal variable names
#    Running this cell will yield errors

9myname = 'Python'
my-name = 'Python'
my name = 'Python'
````

````python
# Multi-word variable naming conventions
myVariableName = 'Python'     # Camel Case:  each word, except the first, 
                              #              starts with a capital letter.
MyVariableName = 'Python'     # Pascal Case: each word starts with a capital 
                              #              letter.
my_variable_name = 'Python'   # Snake Case:  each word is separated by an 
                              #              underscore character.
````

````python
# Combining print() with the `.format()` method.

# NOTE: You can break a line in Python without special notation.
#       In some cases, a formal `\` is required.
#       Indentation of the continued line is inconsequential.

print('My name is {} {}, and I have {} years of Python experience.'.format(
    first, last, n))

print('My name is {one} {two}, and I have {three} years of Python experience.'.
      format(one=first, two=last, three=n))

print('My name is', first + ' ' + last, ', and I have', n,
      'years of Python experience.')
````
